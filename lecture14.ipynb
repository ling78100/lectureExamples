# Plotting with matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

## Histograms

```python
x = np.random.randn(10000)
plt.hist(x, 100)
plt.show()
```

## Plotting

```python
x = range(10)
y = [3 + i + np.random.random() for i in x]
plt.plot(x,y)
plt.show()
```

## Scatter-plot

```python
plt.scatter(x,y)
plt.show()
```

## Multiple lines

```python
y2 = [3 + i + 2*np.random.random() for i in x]
plt.plot(x, y, color='r')
plt.plot(x, y2, color='b')
plt.show()
```

## Axis labeling and legend

```python
y2 = [3 + i + 2*np.random.random() for i in x]
plt.plot(x, y, color='r', label='male')
plt.plot(x, y2, color='b', label='female')
plt.legend()
plt.xlabel("age")
plt.ylabel("income")
plt.title("Legend Example")
plt.show()
```
