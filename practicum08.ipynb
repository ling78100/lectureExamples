```python
#code to get the bigrams
sent="and then we all will see what will happen and then we will know"
sent_split=sent.split()
for i in range(len(sent_split)-1):
    print(sent_split[i:i+2])

for i,w in enumerate(sent_split):
    if i==0: continue
    print(sent_split[i-1],w)
    
```

```python
#using continue for skipping iterations
for a in range(20):
    if a % 2==0 and a !=0: continue
    print(a)
    print(a*2)
    print(a*3)
    print('----')
```

```python
names=["John","Mary","Claire","Sarah","Noah","May","Ana"]
```

```python
#using nested loops for listing all characters in the names list, and skipping vowels
for item in names:
    for ch in item:
        if ch.lower() in "aeiou": continue
        #if ch.lower() in ["a","e","o"]: continue            
        print(item, ch)
```
