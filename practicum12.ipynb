```python
#the HW problem of bad words
#showing how to use regular expression to split words: split_words=re.findall("\w+",text)
#and how to replace only whole words using regular expression: text=re.sub(r"\b%s\b"%word,replacement_word,text)
import re
bad_words="""death
taxes
vanity
in
"""

text="""
Our new Constitution is now established, and has an appearance that 
promises permanency; but in this world nothing can be said to be 
certain, except death and taxes. -Ben Franklin

Indeed I scarce ever heard or saw the introductory Words, Without 
Vanity I may say, etc. but some vain thing immediately follow'd. 
Most People dislike Vanity in others whatever Share they have of 
it themselves, but I give it fair Quarter wherever I meet with it, 
being persuaded that it is often productive of Good to the Possessor 
and to others that are within his Sphere of Action: And therefore in 
many Cases it would not be quite absurd if a Man were to thank God 
for his Vanity among the other deathes Comforts of Life. -Ben Franklin
"""

repl_dict={}
bad_words_list=bad_words.split("\n")
for bw in bad_words_list:
    repl_dict[bw.lower()]="*"*len(bw)

split_words=re.findall("\w+",text)

for w in split_words:
    word_lower=w.lower()
    if word_lower in repl_dict:
        #text=text.replace(w,repl_dict[word_lower])
        text=re.sub(r"\b%s\b"%w,repl_dict[word_lower],text)
        #print(w)
    
print(text)

#print(repl_dict)
#print(split_words)
```

```python
#a code to recursively split a list of numbers into one number, and the rest of numbers as a list, then does the same
#recursively over the rest of numbers
list_numbers=[1,5,6,7,9,10]
def get_largest(input_list,level):
    print("we are working on",input_list,"level",level)
    for n in input_list:
        other_numbers=[v for v in input_list if v!=n]
        print("our number", n, "other numbers", other_numbers)
        get_largest(other_numbers,level+1)

get_largest(list_numbers,0)
```

```python
#a code to recursively generate a full name from a list of names
names=["John","Adam","Smith","Cole","Jack","Mark"]
def create_name(cur_name,level=0):
    if level>3: return
    print(level)
    for name in names:
        #print(name)
        cur_name.append(name)
        new_cur_name=list(cur_name)
        print(new_cur_name)
        create_name(new_cur_name,level+1)
        
create_name([],0)
```
